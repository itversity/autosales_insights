```python
import pandas as pd

# Load the Toyota sales dataset
toyota_sales_data = pd.read_csv("data/car_sales/toyota_sales_data.csv")
```

```python
def calculate_range(series):
    return series.max() - series.min()

custom_aggregations = toyota_sales_data.groupby("sale_rep_id")["sale_amount"]. \
    agg(
        total_sales="sum",
        sales_min="min",
        sales_max="max",
        sales_range=calculate_range
    )
```

```python
custom_aggregations
```

```python
car_model_aggregations = toyota_sales_data.groupby("car_model"). \
    agg(
        total_sales=("sale_amount", "sum"),
        count_sales=("sale_id", "count"),
        sales_range=("sale_amount", calculate_range)
    )
```

```python
car_model_aggregations
```

```python
sales_by_rep_8 = toyota_sales_data.query("sale_rep_id == 8")
```

```python
len(sales_by_rep_8["sale_status"])
```

```python
sum(sales_by_rep_8["sale_status"] == "Completed")
```

```python
def completed_sales_percentage(sale_status_series):
    total_sales = len(sale_status_series)
    completed_sales = (sale_status_series == "Completed").sum()
    return round((completed_sales / total_sales) * 100, 2)
```

```python
completed_sales_percentage(sales_by_rep_8["sale_status"])
```

```python
toyota_sales_data.groupby("sale_rep_id")["sale_status"]. \
    agg(
        total_sales="count",
        completed_pct=completed_sales_percentage
    )
```

```python
toyota_sales_data.groupby("sale_rep_id"). \
    agg(
        total_sales=("sale_amount", "sum"),
        count_sales=("sale_rep_id", "count"),
        completed_sales_pct=("sale_status", completed_sales_percentage)
    )
```
