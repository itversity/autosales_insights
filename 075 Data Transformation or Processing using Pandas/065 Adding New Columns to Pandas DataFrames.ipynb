```python
import pandas as pd

# Load the Toyota sales dataset
toyota_sales_data = pd.read_csv("data/car_sales/toyota_sales_data.csv")
```

```python
toyota_sales_data.head(20)
```

```python
toyota_sales_data["sale_status"] = toyota_sales_data["sale_status"]. \
    replace("Pending", "In Progress")
```

```python
toyota_sales_data.head(20)
```

```python
toyota_sales_data["sale_amount"].max()
```

```python
toyota_sales_data["commission_pct"] = toyota_sales_data["commission_pct"].fillna(0)
```

```python
toyota_sales_data["commission_amount"] = toyota_sales_data["sale_amount"] * toyota_sales_data["commission_pct"]
```

```python
toyota_sales_data["profit_margin"] = (toyota_sales_data["sale_amount"] - toyota_sales_data["commission_amount"]) / toyota_sales_data["sale_amount"]
```

```python
toyota_sales_data.head(20)
```

```python
def get_sale_category(sale):
    if sale["sale_amount"] > 45000:
        return "Very High"
    elif sale["sale_amount"] > 40000:
        return "High"
    elif sale["sale_amount"] > 30000:
        return "Medium"
    else:
        return "Low"
```

```python
toyota_sales_data["sale_category"] = toyota_sales_data.apply(get_sale_category, axis=1)
```
