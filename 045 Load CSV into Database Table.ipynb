```python
sales_reps_df.to_sql(
    'sales_reps_data',
    'postgresql+pg8000://car_sales_user:itversity@localhost:5432/car_sales_db',
    if_exists='append',
    index=False
)
```

```python
from sqlalchemy import create_engine
from sqlalchemy.exc import SQLAlchemyError

engine = create_engine("postgresql+pg8000://car_sales_user:itversity@localhost:5432/car_sales_db", isolation_level="AUTOCOMMIT")

try:
    with engine.connect() as connection:
        with connection.begin():
            sales_reps_df.to_sql(
                'sales_reps_data',
                con=connection,
                if_exists='append',
                index=False
            )
except SQLAlchemyError as e:
    print(f"Database error: {e}")
```
