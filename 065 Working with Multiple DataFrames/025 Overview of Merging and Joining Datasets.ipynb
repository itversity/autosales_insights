## Different types of Joins:
* Inner Join
* Left Join
* Outer Join

## Relating to SQL

| Join Type           | SQL Equivalent             | Pandas `how` Parameter        |
|---------------------|----------------------------|--------------------------------|
| **Left Join**       | `LEFT OUTER JOIN`          | `how='left'`                  |
| **Right Join**      | `RIGHT OUTER JOIN`         | `how='right'`                 |
| **Outer Join**      | `FULL OUTER JOIN`          | `how='outer'`                 |
