```python
import pandas as pd

# Load the Toyota sales dataset
toyota_data = pd.read_csv("data/car_sales/toyota_sales_data.csv")
```

```python
toyota_data["sale_date"] = pd.to_datetime(toyota_data["sale_date"])
```

```python
toyota_data
```

```python
high_sales = toyota_data[toyota_data["sale_amount"] > 30000]
```

```python
high_sales_completed = toyota_data[
    (toyota_data["sale_amount"] > 30000) & (toyota_data["sale_status"] == "Completed")
]
```
