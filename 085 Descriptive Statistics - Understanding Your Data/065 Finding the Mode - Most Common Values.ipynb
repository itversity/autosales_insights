```python
import pandas as pd
```

```python
toyota_sales = pd.read_csv('data/car_sales/toyota_sales_data.csv')
```

```python
toyota_sales["car_model"].value_counts()
```

```python
toyota_sales["car_model"].mode()
```

```python
toyota_sales['sale_status'].value_counts()
```

```python
toyota_sales['commission_pct'].value_counts()
```

```python
toyota_sales['commission_pct'].mode()
```

```python
toyota_sales['sale_amount'].mode()
```

```python
toyota_sales. \
    groupby('sale_rep_id')['car_model']. \
    agg(lambda x: x.mode()[0] if len(x.mode()) > 0 else None). \
    head(20)
```

```python
toyota_sales. \
    groupby(["sale_rep_id", "car_model"])["sale_id"]. \
    count(). \
    reset_index(). \
    rename(columns={"sale_id": "car_count"}). \
    query("sale_rep_id == 10")
```

```python
print("Mean sale amount:", toyota_sales['sale_amount'].mean())
```

```python
print("Median sale amount:", toyota_sales['sale_amount'].median())
```

```python
print("Mode car model:", toyota_sales['car_model'].mode()[0])
```
