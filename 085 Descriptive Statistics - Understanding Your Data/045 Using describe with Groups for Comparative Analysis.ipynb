```python
import pandas as pd
```

```python
toyota_sales = pd.read_csv("data/car_sales/toyota_sales_data.csv")
```

```python
toyota_sales.groupby("car_model")["sale_amount"].describe()
```

```python
sales_reps = pd.read_csv("data/car_sales/sales_reps_data.csv")
```

```python
# Merge to get region info
sales_with_region = toyota_sales.merge(
    sales_reps[['rep_id', 'region']], 
    left_on='sale_rep_id', 
    right_on='rep_id'
)
```

```python
sales_with_region.groupby("region")["sale_amount"].describe()
```

```python
sales_with_region.groupby("region")["sale_amount"].sum()
```
